```python
# Checking the different types of permits in the record to remove permits that are 
# not related to housing construction. These types of permits inlcude 'sign - erect', 
# 'wall or painted sign', 'demolitions', 'grade or quarry or fill or excavate'
print(pd.unique(permits['permit_type_definition']))
print(pd.value_counts(permits['permit_type_definition']))
```

```python
# Define the permit types that need to be removed from the data
permits_2_remove = permits[(permits['permit_type_definition'] == 'sign - erect') | 
                           (permits['permit_type_definition'] == 'wall or painted sign') |
                           (permits['permit_type_definition'] == 'demolitions') |
                           (permits['permit_type_definition'] == 'grade or quarry or fill or excavate')|
                           (permits['permit_type_definition'] == 'otc alterations permit')|
                           (permits['permit_type_definition'] == 'additions alterations or repairs')].index
  
# Drop these given row indexes from dataFrame
permits.drop(permits_2_remove, inplace = True)

# Find and delete permit records that don't have an "Issued Date"
NA_records_Issued = permits[(permits['issued_date'].isna() == True)].index
permits.drop(NA_records_Issued, inplace = True)

# Find and delete permit records that don't have an 'Filed Date'
NA_records_Filed = permits[(permits['filed_date'].isna() == True)].index
permits.drop(NA_records_Filed, inplace = True)
```

```python
#201602179765 2016 Feb 17 withrawn
#permits_101 = permits[permits['permit_number'] == '201602179765']
#201602179758 2016 Feb 17 fire sprinkler
#permits_66 = permits[permits['permit_number'] == '201602179758']
#201602179775 2016 Feb 17 fire spinkler
#permits_30 = permits[permits['permit_number'] == '201602179775']
#201811276751 Fire sprinklers 
#permits_8p1 = permits[permits['permit_number'] == '201602179774']
#201604074211 2016 Feb 17 fire sprinkler
#permits_8p2 = permits[permits['permit_number'] == '201602179774']
#201801088100 
#permits_7 = permits[permits['permit_number'] == '201801088100']

#prop_units_0 = permits[(permits['proposed_units'] == '0')]
#prop_units_0.head(50)

#drop duplicate rows except for the first occurance
permits.drop_duplicates('permit_number', inplace=True)
```

```python
int_columns = ['permit_type', 'number_of_existing_stories', 'number_of_proposed_stories', 'estimated_cost', 'existing_units', 'proposed_units', 'plansets', 'revised_cost']
date_columns = ['status_date', 'filed_date', 'issued_date', 'permit_expiration_date', 'completed_date', 'first_construction_document_date']

for col in int_columns:
  permits[col] = pd.to_numeric(permits[col])

for col in date_columns:
  permits[col] = pd.to_datetime(permits[col])
```

```python
# Create a new column that has the number of days between 'Filed Date' and 'Issued Date' for each permit
permits['wait_time'] = (permits['issued_date'] - permits['filed_date'])/np.timedelta64(1, 'D')
permits.describe()
```

```python
print(pd.value_counts(permits['permit_type_definition']))
```

```python
#print((permits['proposed_units']).sum())
permits['year'] = pd.DatetimeIndex(permits['issued_date']).year
print(pd.DataFrame(permits.groupby('year')['proposed_units'].sum(numeric_only = False)))
permits['year'].isna().sum()
```

```python
plt.hist(permits.proposed_units, bins=50, log=True);
```

```python
#remove all permits with unreasonably high units
permits_units_remove = permits[permits['proposed_units'] >= 800].index

# Drop these given row indexes from dataFrame
permits.drop(permits_units_remove, inplace = True)
```

```python
print((permits['proposed_units']).sum())
```

```python
(permits.isna().sum().sort_values(ascending = False)/len(permits)).plot(kind='bar')
```

```python
#creating a list of all the columns that have greater that 80% NaN's and removing it from the permits dataset
columns_nan = permits.isna().sum().sort_values(ascending = False)/len(permits)
columns_2_remove = columns_nan.loc[columns_nan >= .8].reset_index()
permits = permits.drop(columns=columns_2_remove['index'])
```

```python
#creating Percel Number as a combination of block and lot
permits['Parcel Number'] = permits['block'] + permits['lot']
```

```python
(permits.isna().sum().sort_values(ascending = False)/len(permits)).plot(kind='bar')
```

```python
#site permit columns is a bolean that is filled with 'Y' or None. We need to fill it out with a bolean True of False
permits = permits.replace({'site_permit': {'Y': True, None: False}})
```
