```python
sftax = pd.read_csv('C:/Users/Rohan/Documents/NYU/PUI/Assessor_Historical_Secured_Property_Tax_Rolls.csv',\
                usecols=['Use Code','Use Definition','Property Class Code','Property Class Code Definition',
                         'Closed Roll Year', 'Property Location', 'Parcel Number', 'Block',
       'Lot', 'Year Property Built', 'Number of Bathrooms', 'Number of Bedrooms',
       'Number of Rooms', 'Number of Stories', 'Number of Units','Status Code',
       'Zoning Code', 'Construction Type', 'Lot Depth', 'Lot Frontage',
       'Property Area', 'Basement Area', 'Lot Area', 
       'Tax Rate Area Code', 'Percent of Ownership', 'Homeowner Exemption Value', 'Current Sales Date',
       'Assessed Fixtures Value', 'Assessed Improvement Value',
       'Assessed Land Value', 'Assessed Personal Property Value',
       'the_geom', 'Row ID']); sftax.head()
```

```python
sftax.isna().sum().sort_values(ascending = False)
```

```python
sftax_num = sftax.select_dtypes(include = [np.number])
sftax_num_cols = sftax_num.columns.values
print(sftax_num_cols)
```

```python
sftax_nonnum = sftax.select_dtypes(exclude = [np.number])
sftax_nonnum_cols = sftax_nonnum.columns.values
print(sftax_nonnum_cols)
```

```python
#permit data is >2013, so filter out parcels with assessments before 2013
sftax_2013 = sftax.loc[sftax['Closed Roll Year'] >= 2013]; (sftax.shape, sftax_2013.shape)
```

```python
(sftax_2013.duplicated(subset = 'Parcel Number').sum())/len(sftax_2013)
```

```python
highest_zero_cols = list(((sftax_2013 == 0).sum(axis = 0)/len(sftax_2013)).sort_values(ascending = False)[:6].index)
sftax_2013 = sftax_2013.drop(labels = highest_zero_cols, axis = 'columns')
sftax_2013.head()
```

```python
sftax_2013['Total Taxable Value'] = sftax_2013['Assessed Land Value'] + sftax_2013['Assessed Improvement Value'] - sftax_2013['Homeowner Exemption Value']
sftax_2013['Total Taxable Value'].describe()
```

```python
sftax_2013.isna().sum().sort_values(ascending = False)
```

```python
sftax_2013 = sftax_2013.drop(labels = ['Status Code','Current Sales Date','Construction Type','Zoning Code',
                                      'Year Property Built','Tax Rate Area Code','the_geom'],
                             axis = 'columns')
sftax_2013.columns
```

```python
sftax_2013 = sftax_2013.dropna(axis = 'index', subset = ['Use Code', 'Property Class Code',
                                            'Total Taxable Value', 'Number of Rooms'])
sftax_2013.isna().sum().sort_values(ascending = False)[:8]
```

```python
sftax_2013_regs = sftax_2013.loc[:,['Closed Roll Year','Parcel Number','Number of Rooms',
                                  'Property Area','Total Taxable Value', 'Use Code','Property Class Code']]; 
sftax_2013_regs.head()
```

```python
((sftax_2013_regs == 0).sum(axis = 0)/len(sftax_2013_regs)).sort_values(ascending = False)
#sanityIndex = (sftax_2013_regs['Number of Rooms'] > 0) & (sftax_2013_regs['Property Area'] > 0) 
#sftax_2013_regs = sftax_2013_regs[sanityIndex]
```

```python
outlierIndex = (sftax_2013_regs['Total Taxable Value'] < sftax_2013_regs['Total Taxable Value'].quantile(0.9999)) 
#& (sftax_2013_regs['Property Area'] > sftax_2013_regs['Property Area'].quantile(0.001)) & \
#(sftax_2013_regs['Property Area'] < sftax_2013_regs['Property Area'].quantile(0.999)) & \
#(sftax_2013_regs['Number of Rooms'] < sftax_2013_regs['Number of Rooms'].quantile(0.999)) 
sftax_2013_regs = sftax_2013_regs[outlierIndex]
sftax_2013_regs.describe()
```

```python
#separate out all use/class codes for parcels that have residential or mixed uses
RES_mixeduse_index = (sftax_2013_regs['Use Code'] == 'MRES') | (sftax_2013_regs['Use Code'] == 'SRES') | 
(sftax_2013_regs['Property Class Code'] == 'AC') | (sftax_2013_regs['Property Class Code'] == 'CM') | 
(sftax_2013_regs['Property Class Code'] == 'CZ') | (sftax_2013_regs['Property Class Code'] == 'MB') | 
(sftax_2013_regs['Property Class Code'] == 'MIX') | (sftax_2013_regs['Property Class Code'] == 'RDAP') | 
(sftax_2013_regs['Property Class Code'] == 'VA15') | (sftax_2013_regs['Property Class Code'] == 'VRX')
                          
RES_mixeduse_parcels = sftax_2013_regs.loc[RES_mixeduse_index]

APN_point_estimate_regs = RES_mixeduse_parcels.groupby(by = 'Parcel Number', as_index = False).agg(
{'Number of Rooms':'mean','Property Area':'mean','Total Taxable Value':'median'})

APN_point_estimate_regs
```
